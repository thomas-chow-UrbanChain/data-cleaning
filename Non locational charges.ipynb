```python
import pandas as pd
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings("ignore")

df = pd.read_excel('5-Year TNUoS Tariff Report Tables - Apr24v3_External 1.xlsx', sheet_name = 'TB', skiprows = [0, 1, 2])
years = [{'Domestic':0.104586, 'LV_NoMIC_1':0.069796, 'LV_NoMIC_2':0.253213, 'LV_NoMIC_3':0.584795, 'LV_NoMIC_4':1.741054, 'LV1':3.129643, 'LV2':5.323905, 'LV3':8.488176, 'LV4':19.793806, 'HV1':16.455194, 'HV2':49.655569, 'HV3':95.286086, 'HV4':243.634626, 'EHV1':132.851337, 'EHV2':668.538064, 'EHV3':1255.850, 'EHV4':3520.063002, 'T-Demand1':397.072685, 'T-Demand2':1611.510222, 'T-Demand3':3754.664746, 'T-Demand4':11958.120683, 'Unmetered':1.188571}]
periods = ['2025/26', '2026/27', '2027/28', '2028/29']
for p in periods:
    years.append(pd.Series(df[p].iloc[1:24].drop([22]).values,index=df.Band.iloc[1:24].drop([22])).to_dict())
```

```python
output = pd.DataFrame(columns = list(years[0].keys()))
output
```

```python

# Split start date to convert to datetime
start_date = datetime(2024,4,1,0,0,0)
end_date = datetime(2029,3,31,23,30,0)

dates_range = []
# Increment every 30 minutes from start date to end date
while start_date <= end_date:
    dates_range.append(start_date)
    start_date += timedelta(minutes=30)
output['Date (UTC)'] = dates_range
```

```python
def convert(time):
    if time < datetime(2025,4,1,0,0,0):
        return 0
    elif time < datetime(2026,4,1,0,0,0):
        return 1
    elif time < datetime(2027,4,1,0,0,0):
        return 2
    elif time < datetime(2028,4,1,0,0,0):
        return 3
    else:
        return 4
```

```python
for c in output.columns[:-1]:
    output[c] = output['Date (UTC)'].apply(lambda x: years[convert(x)][c])
```

```python
col = list(output.columns[:-1].insert(0,'Date (UTC)'))
output = output[col]
output
```

```python
output.to_csv('non-locational charges.csv', index = False)
```
