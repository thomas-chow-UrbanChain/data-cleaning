```python
import pandas as pd
import pytz
from datetime import datetime, timedelta
df = pd.DataFrame(columns = ['Charging Date/Time UTC', 'Charging Date/Time  Clock', 'Probability',
       'Northern Scotland', 'Southern Scotland', 'Northern', 'North West',
       'Yorkshire', 'N Wales & Mersey', 'East Midlands', 'Midlands', 'Eastern',
       'South Wales', 'South East', 'London', 'Southern', 'South Western'])
df
```

```python
# Split start date to convert to datetime
start_date = datetime(2024,3,31,23,0,0)
end_date = datetime(2029,3,31,22,30,0)

utc_dates = []
clock_dates = []
# Increment every 30 minutes from start date to end date
while start_date <= end_date:
    utc_dates.append(start_date)
    
    tz = pytz.timezone('UTC')
    tz_time = tz.localize(start_date)
    london_tz = pytz.timezone('Europe/London')
    london_time = tz_time.astimezone(london_tz)
    clock_dates.append(london_time.replace(tzinfo=None))
    
    start_date += timedelta(minutes=30)
df['Charging Date/Time UTC'] = utc_dates
df['Charging Date/Time  Clock'] = clock_dates
```

```python
map = pd.read_excel('5-Year TNUoS Tariff Report Tables - Apr24v3_External 1.xlsx', sheet_name = 'T19 & Fig 6', skiprows = [0,1])
map
```

```python
years = [{'Northern Scotland': 0.0,'Southern Scotland': 0.0,'Northern': 0.0,'North West': 0.0,'Yorkshire': 0.0,'N Wales & Mersey': 0.0,'East Midlands': 0.0,'Midlands':0.312657,
  'Eastern': 0.113949,'South Wales': 0.533793,'South East': 0.538522,'London': 0.644217,'Southern': 0.903934,'South Western': 1.129620}]
periods = ['2025/26 (p/kWh)', '2026/27 (p/kWh)', '2027/28 (p/kWh)', '2028/29 (p/kWh)']
for p in periods:
    years.append(pd.Series(map[p].iloc[0:14].values,index=map['Zone Name'].iloc[0:14]).to_dict())
```

```python
def convert(time):
    if time < datetime(2025,4,1,0,0,0):
        return 0
    elif time < datetime(2026,4,1,0,0,0):
        return 1
    elif time < datetime(2027,4,1,0,0,0):
        return 2
    elif time < datetime(2028,4,1,0,0,0):
        return 3
    else:
        return 4
```

```python
for c in df.columns[3:]:
    df[c] = df['Charging Date/Time  Clock'].apply(lambda x: years[convert(x)][c])
df
```

```python
df.to_csv('NHH demand tariff.csv', index = False)
```
